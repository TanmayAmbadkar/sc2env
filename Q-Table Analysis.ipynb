```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the Q-table from the CSV file
q_table = pd.read_csv('../Q_table_m_e_1.csv', index_col=0)
q_table.head()
```

```python
# Display basic statistics
q_table.describe()
```

```python
plt.figure(figsize=(10, 8))
sns.heatmap(q_table, annot=True, cmap='coolwarm')
plt.title('Heatmap of Q-Table')
plt.xlabel('Actions')
plt.ylabel('States')
plt.show()
```
